```python
#checking the onsets and time presentations 
import pandas as pd
import pickle
import numpy as np
#import matplotlib.pyplot as plt

unpickled_df = pd.read_pickle("/Users/nestor/Documents/Marburg/FRAPPS/logfiles/FaceID/logfile_s001.pkl")
unpickled_df

data = unpickled_df
del data['subjectID'] #deleting some unnecessary info
del data['answers order'] 

df = pd.DataFrame.from_dict(data)
df
#df.to_csv("C:\\Users\\Nestor\\Desktop\\FRAPPS\\behv_data\\out\\sub-001\\DataFrame_OnsetsNallevents.csv")
```

```python
#extracting the onsets and durations of the events. Face, questions , button pressed 
#FACE ONSETS
OnstsNfacs = df.iloc[::5, :]
OnstsNfacs.head()
```

```python
OnstsNfacs.rename(columns={"event" : "Faces"})
```

```python
Onsts4faces = OnstsNfacs[['onset']]
Onsts4faces.head()
```

```python
Onsts4faces.rename(columns={"onset" : "facesTime"})
```

```python
#QUESTIONS ONSETS
QuestionOnsets = df[df.index % 5 == 1]
QuestionOnsets.head()
```

```python
QuestionOnsets.rename(columns={"event" : "Questions"})
```

```python
Onsts4Questions = QuestionOnsets[['onset']]
Onsts4Questions.head()
```

```python
Onsts4Questions.rename(columns={"onset" : "QuestionT"})
```

```python
#Button pressed ONSETS
BttnPrss = df[df.index % 5 == 2]
BttnPrss.head()
```

```python
Onsts4button = BttnPrss[['onset']]
Onsts4button.head()
```

```python
Onsts4button.rename(columns={"onset" : "ButtonT"})
```

```python
TrialRsltOnst = df[df.index % 5 == 4]
TrialRsltOnst.head()
```

```python
NInd_Onsts4faces = Onsts4faces.reset_index(drop=True)
NInd_Onsts4Questions = Onsts4Questions.reset_index(drop=True)
NInd_Onsts4button = Onsts4button.reset_index(drop=True)
```

```python
Onsets = pd.concat([NInd_Onsts4faces, NInd_Onsts4Questions, NInd_Onsts4button], axis=1, keys=['facesTime', 'QuestionT', 'ButtonT'], ignore_index=False)
Onsets
```
