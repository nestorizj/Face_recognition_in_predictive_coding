```python
# y-axis in bold
rc('font', weight='bold')
header = ['yes', 'no', 'missed']
 
# Values of each group
yes = [119,149,140]
no = [69, 40, 44] 
missed = [1, 0, 5]
 
# Heights of bars1 + bars2
bars = np.add(yes, no).tolist()
 
# The position of the bars on the x-axis
r = [0,1,2]
 
# Names of group and bar width
names = ['sub1','sub2','sub3']
barWidth = 0.35
 
# Create brown bars
p1 = plt.bar(r, yes, color='r', edgecolor='white', width=barWidth)
# Create green bars (middle), on top of the firs ones
p2 = plt.bar(r, no, bottom = yes, color ='g', edgecolor='white', width=barWidth)
# Create green bars (top)
p3 = plt.bar(r, missed, bottom=bars, color='b', edgecolor='white', width=barWidth)
 
# Custom X axis
plt.xticks(r, names, fontweight='bold')
plt.xlabel('participants')

# Custom Y axis
plt.ylabel ('scores')
plt.yticks(np.arange(0,250,10))
plt.legend((p1[0], p2[0], p3[0]), (header[0], header[1], header[2]))

# Show graphic
plt.show()
```
